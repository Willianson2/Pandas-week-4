```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
#modul yang dipakai
```

```python
n_rows = 10
n_cols = 1
cols = ('usia',)

df = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)),
                  columns=cols)
               
df
#persiapan data frame
#usia yang dimiliki bisa berbeda karaena lewat proses random
```

```python
df['kelompok_usia'] = pd.cut(df['usia'],
                            bins = [0, 18, 65, 99],
                            labels = ['anak', 'dewasa', 'manula']) 
df
#mengelompokkan usia anak, dewasa, manula menggunakan method cut
#proses pemotongan menggunakan bins    
#anak anak 0-18 dewasa 18-65 manula 65-99
```
